````python
def licenseKeyFormatting(s, k):
  """transform the input string (only letters, digits, and '-') into groups of string
  get rid of '-'s,group size is k, there are '-' between groups, the first group may be k, or less than k"""
  count = 0 # need a counter
  results = [] # the result string will be returned.
  # starting from the last character, and in reversed order
  for i in reversed(range(len(s))):
    # check if the character is '-', if not, uppercase it and put into the result string, add 1 to count
    if s[i] != '-':

      results.append(s[i].upper())
      count = count + 1

      # check if count is equal to group size k, if yes, count reset to zero, the result string add '-'
      if count == k:
        count = 0
        results.append('-')

  if (len(s) > 0 and results[-1] == '-'):  # check if the last character is '-'
    results = results[:-1] # if it is, remove it
  results = results[::-1]   # reverse the whole list
  results = "".join(results) # make a whole list into a string using .join() function
  return results

            
````

### [Problem 383](https://leetcode.com/problems/ransom-note/) - Ransom Note

Given two strings `ransomNote` and `magazine`, return `true` if `ransomNote` can be constructed by using the letters from `magazine` and `false` otherwise.

Each letter in `magazine` can only be used once in `ransomNote`.

**Example 1:**
```
Input: ransomNote = "a", magazine = "b"
Output: false
```

**Example 2:**
```
Input: ransomNote = "aa", magazine = "ab"
Output: false
```

**Example 3:**
```
Input: ransomNote = "aa", magazine = "aab"
Output: true
```

**Constraints:**
```
1 <= ransomNote.length, magazine.length <= 105
ransomNote and magazine consist of lowercase English letters.
```

````python
def count_occurrences(s):
  """compute the occurrence of each letter in the given string, and returns a dictionary with
  letter as key, occurrences as value"""
  counts = {}
  for letter in s:
    if letter in counts:
      # if this letter has already been in the dictionary, add 1 to count 
      counts[letter] = counts[letter] + 1
    else:
      # if not present already, create an entry
      counts[letter] = 1
  return counts
````

````python
count_occurrences('aab')
````

````python
def canConstruct(s1,s2):
  # create an occurrences dictionary for each string
  d1 = count_occurrences(s1)
  print("d1", d1)
  d2 = count_occurrences(s2)
  print("d2", d2)

  flag = 0
  for letter in d1:
    for character in d2:
      if letter == character and d1[letter] <= d2[character]:
         flag += 1
  if flag == len(d1):
    return True
  else:
    return False
````
